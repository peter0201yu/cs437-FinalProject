````python
genres = ['classical', 'club', 'country', 'dance', 'disco', \
          'edm', 'gospel', 'guitar', 'hard-rock', 'hip-hop', 'house', \
          'indie-pop', 'jazz', 'k-pop', 'latin', 'piano', 'pop', \
          'punk', 'r-n-b', 'rainy-day', 'rock', 'rock-n-roll', \
          'salsa', 'soul', 'soundtrack', 'study', 'workout' \
          'world-music']
````

````python
tracks = sp.recommendations(seed_genres=['classical'], limit=50, min_popularity=50)
print(tracks['tracks'][0].keys())
print(tracks['tracks'][0])
````

````python
artist_id = []
track_name = []
popularity = []
track_id = []
genre_id = []
urls = []

for i, genre in enumerate(genres):
    tracks = sp.recommendations(seed_genres=[genre], limit=50, min_popularity=50)
    for t in tracks['tracks']:
        track_id.append(t['id'])
        artist_id.append(t['artists'][0]['id'])
        track_name.append(t['name'])
        genre_id.append(i)
        popularity.append(t['popularity'])
        urls.append(t['external_urls']['spotify'])
````

````python
import pandas as pd
track_dataframe = pd.DataFrame({'track_id' : track_id, 'track_name' : track_name, 'artist_id' : artist_id, 'genre_id' : genre_id, 'popularity' : popularity, 'url' : urls})
track_dataframe = track_dataframe.drop_duplicates(subset='track_id', keep='first')
print(track_dataframe.shape)
track_dataframe.head()
````

````python
pop_genre = track_dataframe[track_dataframe['genre_id'] == genres.index('pop')]

# Displaying the first five rows where genre is 'pop'
pop_genre.head()
````

````python
dtype = {
    'track_id': types.VARCHAR(length=32),
    'artist_id': types.VARCHAR(length=32),
    'track_name': types.VARCHAR(length=256),
    'genre_id': types.INTEGER, 
    'popularity': types.INTEGER,
    'url': types.VARCHAR(length=128)
}
track_dataframe.to_sql('songs', engine, if_exists='replace', index=False, dtype=dtype)
````

````python
artist_id_list = []
artist_name = []
artist_genre = []

# artists = sp.artist(list(set(artist_id)))
for aid in set(artist_id):
    artist = sp.artist(aid)
    artist_id_list.append(artist['id'])
    artist_name.append(artist['name'])
    artist_genre.append(artist['genres'])

artist_dataframe = pd.DataFrame({'artist_id' : artist_id_list, 'artist_name' : artist_name, 'genre_id' : artist_genre})
print(artist_dataframe.shape)
artist_dataframe.head()
````

````python
# process the genres
cleaned_genres_ids = []
genres[17]
for genre_names in artist_genre:
    gid = -1
    for genre_name in genre_names:
        l = genre_name.split(" ")
        found = False
        for name in l:
            if name in genres:
                gid = genres.index(name)
                found = True
                break
            if name == 'r&b':
                gid = 17 # written as r-n-b in the list
                found = True
                break
            if name == 'hip' or name == 'rap':
                gid = 9
                found=True
                break
        if found:
            break
    cleaned_genres_ids.append(gid)
````

````python
artist_dataframe = pd.DataFrame({'artist_id' : artist_id_list, 'artist_name' : artist_name, 'genre_id' : cleaned_genres_ids})
artist_dataframe.head()
````

````python
artist_id = []
track_name = []
track_id = []
urls = []
url_prefix = "https://music.youtube.com/watch?v="
setting_list = []
for artist in popular_artists:
    aid = artist_dataframe.loc[artist_dataframe['artist_name'] == artist, 'artist_id'].squeeze()
    print(aid)
    for setting in settings:
        search_results = yt.search("{} {}".format(artist, setting), filter='videos', limit=3)
        for res in search_results[:3]:
            track_id.append(res['videoId'])
            track_name.append(res['title'])
            artist_id.append(aid)
            urls.append(url_prefix + res['videoId'])
            setting_list.append(settings.index(setting))

yt_track_dataframe = pd.DataFrame({'track_id' : track_id, 'track_name' : track_name, 'artist_id' : artist_id, 'setting_id': setting_list, 'url' : urls})
yt_track_dataframe = yt_track_dataframe.drop_duplicates(subset=['track_id'])
yt_track_dataframe.head(10)
````
